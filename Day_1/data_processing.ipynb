```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Read the data
df = pd.read_csv("sample_data.csv")

# 1. Apply Label Encoding
df_label = df.copy()
le = LabelEncoder()

df_label['Gender_Encoded'] = le.fit_transform(df_label['Gender'])
df_label['Passed_Encoded'] = le.fit_transform(df_label['Passed'])

print('\n--- LabelEncoder Data ---')
print(df_label[['Name', 'Gender', 'Gender_Encoded', 'Passed', 'Passed_Encoded']])

# 2. Apply One-Hot Encoding (Using Option 2 to keep it clean)
# This will turn 'City' into binary columns (True/False or 1/0)
df_encoded = pd.get_dummies(df_label, columns=['City'] , dtype=int)

print('\n--- Final Dataframe with One-Hot Encoded City ---')
print(df_encoded.head())
```
